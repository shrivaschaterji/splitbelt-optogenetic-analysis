```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.ndimage import gaussian_filter1d
import glob
import os
import re
```

```python
t = np.linspace(0, 60, 6001)  # 0 to 60s in 0.01s steps
signal = np.zeros_like(t)     # Initialize signal with zeros

# Turn signal ON for 2 seconds every 10 seconds starting at 5s
onsets = np.arange(5, 60, 10)  # 5, 15, 25, ..., 55

for onset in onsets:
    signal[(t >= onset) & (t < onset + 2)] = 1

# Plot
plt.figure(figsize=(10, 2))
plt.plot(t, signal, drawstyle='steps-post')
plt.xlabel('Time (s)')
plt.ylabel('Laser Signal')
plt.title('Laser Signal Schematic')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

path = 'C:\\Users\\User\\Carey Lab Dropbox\\Rotation Carey\\Shrivas and Coralie\DATA\RN-BIP-JAWS-INHIBITION-2s-4sIPI'
csv_files = glob.glob(os.path.join(path, "*1000000.csv"))

def extract_animal_and_trial(filepath):
    filename = os.path.basename(filepath)
    
    # Match animal name: everything before first underscore
    animal_match = re.match(r'([A-Za-z0-9]+)_', filename)
    animal = animal_match.group(1) if animal_match else ""

    # Match trial number before DLC
    trial_match = re.search(r'_([0-9]+)DLC', filename)
    trial = int(trial_match.group(1)) if trial_match else float('inf')

    return (animal, trial)

# Sort using both animal and trial
csv_files.sort(key=extract_animal_and_trial)
for file in csv_files:
    condition = 'Inhibition'  # Extract condition from filename
    print(condition)
    animal = os.path.basename(file).split('_')[0]  # Extract animal name from filename
    print(animal)
    trial = os.path.basename(file).split('_')[7]  # Extract trial number from filename
    trial =  re.findall(r'\d+', trial)[0]  # Extract the first number from the string
    print(trial)
    strideFile = pd.read_csv(file, header=1, encoding='unicode_escape')

    # Extract FRside X values (removing the first row)
    FRbX = strideFile["FRbottom"]
    FRbX = FRbX.tail(-1).astype(float).reset_index(drop=True)
    FLbX = strideFile["FLbottom"]
    FLbX = FLbX.tail(-1).astype(float).reset_index(drop=True)
    BCbX = strideFile["bodycenterbottom"]
    BCbX = BCbX.tail(-1).astype(float).reset_index(drop=True)
    

    FRrelX = FRbX - BCbX  # Calculate relative position

    FLrelX = FLbX - BCbX
    sigma = 10  # Standard deviation for Gaussian kernel (adjust for more/less smoothing)
    FRrelX = gaussian_filter1d(FRrelX, sigma=sigma)
    FLrelX = gaussian_filter1d(FLrelX, sigma=sigma)
    # Time vector for 33 Hz sampling
    frame_rate = 333  # Hz

    t = np.arange(len(FRbX))/ frame_rate  # Time in seconds

    # --- Generate Laser Signal ---
    signal = np.zeros_like(t)
    onsets = np.arange(5, 60, 10)  # Laser turns on at 5, 15, 25... for 2 seconds

    for onset in onsets:
        signal[(t >= onset) & (t < onset + 2)] = 1

    # --- Plot with Black Background ---
    fig, ax1 = plt.subplots(figsize=(25, 4), facecolor='black')
    ax1.set_facecolor('black')

    # Plot FR X-position
    ax1.plot(t, FRrelX, label='FRbottom X', color='lightcoral')
    ax1.plot(t, FLrelX, label='FLbottom X', color='cyan', alpha=0.5)
    ax1.set_xlabel('Time (s)', color='white')
    ax1.set_ylabel('FPawsbottom X Position', color='white')
    ax1.tick_params(axis='y', labelcolor='white')
    ax1.tick_params(axis='x', colors='white')
    ax1.set_ylim(-50, 200)

    # Add secondary Y-axis for laser signal
    ax2 = ax1.twinx()
    ax2.plot(t, signal, drawstyle='steps-post', color='yellow', label='Laser Signal')
    ax2.set_ylabel('Laser', color='yellow')
    ax2.tick_params(axis='y', labelcolor='yellow')
    ax2.set_ylim(-0.1, 1.1)

    # Customize titles and grid
    plt.suptitle('FPawsbottom X-Position with Laser Signal Overlay', color='white')
    plt.title(f'{animal}, Trial {trial}, {condition}', color='white')
    ax1.grid(False)
    fig.tight_layout()


    #plt.savefig('FRside_X_Position_with_Laser_Signal_Overlay.png', facecolor='black')
    plt.show()
```
